### PCA

Principal Component Analysis (PCA) is a technique widely used in signal processing to simplify and extract relevant information from complex data sets. In signal processing, PCA helps in reducing the dimensionality of the data while retaining the most important characteristics.

```python
def PCA_component(data_matrix, num_components=1):
    """
    Get the most important principal component(s) of a data matrix using PCA.

    Parameters:
      data_matrix (array-like): Input data matrix with shape (number of samples, number of features).
      num_components (int): Number of principal components to extract, default is 1.

    Returns:
      most_important_component (array): The most important principal component(s) with shape (number of features,).

    Usage examples:
      import numpy as np

      # 4 samples and 3 features
      data_matrix = np.random.rand(4, 3)
      num_components = 2
      most_important_pc = PCA_component(data_matrix, num_components)

      # Print principal components
      print("Most important principal component:", most_important_pc)
    """

    # Create a PCA model
    pca = PCA(n_components=num_components)

    # Fit the data and perform PCA transformation
    pca.fit(data_matrix)
    transformed_data = pca.transform(data_matrix)

    # Get the most important principal component(s) (first principal component)
    # print(pca.components_)
    PCA_components = pca.components_[:num_components]

    return PCA_components
```
