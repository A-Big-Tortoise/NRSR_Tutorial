#### Haar Wavelet
低通分解滤波器 h[n], 对于方形Haar小波，我们只需在t = 0和t = 1时采样其值：

$$ h[n] = \begin{cases}
\frac{1}{\sqrt{2}} & \text{if } n = 0 \\
\frac{1}{\sqrt{2}} & \text{if } n = 1 \\
0 & \text{otherwise}
\end{cases} $$

高通分解滤波器 g[n]：

$$ g[n] = \begin{cases}
\frac{1}{\sqrt{2}} & \text{if } n = 0 \\
-\frac{1}{\sqrt{2}} & \text{if } n = 1 \\
0 & \text{otherwise}
\end{cases} $$

除以根号2是为了满足单位能量性质，确保小波函数具有单位总能量。

```python
def haar_wavelet(t):
    if 0 <= t < 0.5:
        return 1
    elif 0.5 <= t < 1:
        return -1
    else:
        return 0

t = np.linspace(-2, 2, num=200)

haar_values = [haar_wavelet(ti) for ti in t]

plt.plot(t, haar_values, label='Haar Wavelet', color='blue')
plt.xlabel('Time')
plt.ylabel('Value of Haar Wavelet')
plt.title('Haar Wavelet')
plt.grid(True)
plt.legend()
plt.show()
```


#### Daubechies Wavelet
**紧支持（Compact Support）**：
它们在时域上在有限的时间范围内非零，而在其他时间段内为零。紧支持性质使Daubechies小波在分析瞬态信号、突发事件或非平稳信号时非常有效，因为它们不会受到信号整体特性的影响。


**多分辨率（Multiresolution）**：
Daubechies小波可对信号的不同频率成分进行多层次的分解和重构。在小波分析中，我们可以选择不同尺度的小波函数来分析信号，从而在不同层次上细致地观察信号的频率成分。


**紧包性（Tight Frame）**：
紧包表示它们的小波系数之和等于信号的能量。确保了在信号分解和重构的过程中不会引入额外的能量或信息丢失。紧包性质保持了小波变换的正交性以及信号的完整性。

```python
wavelet = 'db8'
wavelet_obj = pywt.Wavelet(wavelet)

lpf_dec = wavelet_obj.dec_lo

plt.stem(np.arange(len(lpf_dec)), lpf_dec, basefmt='b-')
plt.title('Daubechies-8 Low-Pass Decomposition Filter Coefficients')
plt.xlabel('Sampling Point')
plt.ylabel('Filter Coefficients')
plt.grid(True)
plt.show()
```


#### Coiflet Wavelet
**对称性**：
Coiflet小波是对称的。这意味着在频域中也是对称的。对称性有助于保持信号的对称性特征。


**高阶多项式逼近**：
Coiflet小波在时域上是高阶多项式的逼近，这意味着它们可以更好地逼近信号中的高阶趋势。这使得Coiflet小波在处理信号的高频细节时表现出色，因为能够更好地捕获信号中的高次变化。

```python
wavelet = 'coif8'
wavelet_obj = pywt.Wavelet(wavelet)

lpf_dec = wavelet_obj.dec_lo

plt.stem(np.arange(len(lpf_dec)), lpf_dec, basefmt='b-')
plt.title('Coiflet-8 Low-Pass Decomposition Filter Coefficients')
plt.xlabel('Sampling Point')
plt.ylabel('Filter Coefficients Value')
plt.grid(True)
plt.show()
```


#### Symlet Wavelet

**可调节的对称性**：
Symlet小波的对称性是可调节的，它们可以具有不同阶数的对称性。这种可变对称性允许Symlet小波可以是对称的、反对称的，或介于两者之间。

**频率局部化**：
Symlet小波在频域上具有较好的局部化特性，能够将信号的频率特征更精确地捕获。这种频率局部化性质有助于在信号分析中准确地定位和识别不同频率成分。

```python
wavelet = 'sym8'
wavelet_obj = pywt.Wavelet(wavelet)

lpf_dec = wavelet_obj.dec_lo

plt.stem(np.arange(len(lpf_dec)), lpf_dec, basefmt='b-')
plt.title('Symlet-8 Low-Pass Decomposition Filter Coefficients')
plt.xlabel('Sampling Point')
plt.ylabel('Filter Coefficients Value')
plt.grid(True)
plt.show()
```
