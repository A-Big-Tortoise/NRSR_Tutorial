#### A Playground to try  Discrete Wavelet Transform(DWT) and Empirical Wavelet Transform(EWT) on SCG Signals

```python
from Code.Tutorial.Dataset import load_scg

import numpy as np
import matplotlib.pyplot as plt
import pywt
```

```python
def plot_dwt(signal, coeffs, level):
    # Create a time sequence
    time = np.arange(0, len(signal))

    # Plot the original signal
    plt.figure(figsize=(12, 2.5 * (level+2)))
    plt.subplot(level+2, 1, 1)
    plt.plot(time, signal, 'b', label='Original Signal')
    plt.title('Original Signal')
    plt.xlabel('Time')
    plt.ylabel('Amplitude')
    plt.grid()

    # Plot the results of wavelet transform
    for i, coeff in enumerate(coeffs):
        plt.subplot(level+2, 1, i+2)
        plt.plot(time[:len(coeff)], coeff, label=f'Level {i+1}')
        plt.title(f'Wavelet Transform Result - Level {i+1}')
        plt.xlabel('Time')
        plt.ylabel('Amplitude')
        plt.legend()
        plt.grid()

    plt.tight_layout()
    plt.show()
```

```python
# load data
signals, labels, duration, fs = load_scg(0, 'train')
signal = signals[0]

wavelet = 'db4'  # Choose the wavelet function, using Daubechies 4 wavelet here
level = 3 # Set the level of wavelet transform
coeffs = pywt.wavedec(signal, wavelet, level=level)
plot_dwt(signal, coeffs, level)
```
