```python
import numpy as np
import matplotlib.pyplot as plt
from Core2dsp import PLOT_Jiayu
from NRSR_Tutorial.dsp import scg_simulate, cal_stft, cal_fft, cal_gabor

scg_label = scg_simulate(n_echo=0)
scg, label = scg_label[:1000], scg_label[-6:]
PLOT_Jiayu.plot(scg)
```

```python
from scipy.fftpack import dct, rfft
from NRSR_Tutorial.dsp import sine_wave, add_distort_noise
# signal = scg
signal = scg[:int(6000//label[2])]
duration = 2
signal = sine_wave(duration=duration, sampling_rate=100, amplitude=1, frequency=1.1) + sine_wave(duration=duration, sampling_rate=100, amplitude=2, frequency=10.1) + sine_wave(duration=duration, sampling_rate=100, amplitude=5, frequency=3.1)
# signal = add_distort_noise(signal, noise_frequency=30, n_samples=duration*100, noise_amplitude=.3)
# print(scg_label)
plt.plot(signal)
plt.show()

dct_res = dct(signal, type=2)
plt.stem(np.abs(dct_res))
plt.show()
plt.stem(np.abs(rfft(signal)))
plt.show()

plt.plot(np.cumsum(np.square(np.abs(dct_res)))/np.sum(np.square(np.abs(dct_res))), label='dct')
plt.plot(np.cumsum(np.square(np.abs(rfft(signal))))/np.sum(np.square(np.abs(rfft(signal)))), label='dft')
plt.legend()
```
